```python
from chembl_webresource_client.new_client import new_client
import pandas as pd

def get_chembl_data(target_name, max_compounds=1000):

    target = new_client.target
    res = target.search(target_name)

    for i, t in enumerate(res):
        print(i, t['target_chembl_id'], t['pref_name'], t['organism'])
    target_id = None
    for t in res:
        if t['pref_name'] == "Epidermal growth factor receptor" and t['organism'] == "Homo sapiens":
            target_id = t['target_chembl_id']
            break



    activity = new_client.activity
    acts = activity.filter(
        target_chembl_id=target_id,
        standard_type='IC50',
        relation='=',
        assay_type='B'
    ).only([
        'molecule_chembl_id',
        'canonical_smiles',
        'standard_value',
        'standard_units',
        'standard_type',
        'assay_chembl_id'
    ])

    df = pd.DataFrame(list(acts))
    print(f"number of molecules: {len(df)}")

    return df
```
